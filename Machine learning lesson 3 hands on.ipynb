```python
Titanic.drop('alive', axis = 1, inplace = True)
```

```python
Titanic.drop('alone', axis = 1, inplace = True)
Titanic.drop('embark_town', axis = 1, inplace = True)
Titanic.head()
```

```python
Titanic.info()
```

```python
Titanic.age = Titanic.age.astype(int)
Titanic.fare = Titanic.fare.astype(int)
Titanic.info()
```

```python
Titanic.deck.value_counts()
```

```python
def Deck(series):
    if series == 'A':
        return 0
    if series == 'B':
        return 1
    if series == 'C':
        return 2
    if series == 'D':
        return 3
    if series == 'E':
        return 4
    if series == 'F':
        return 5 
    if series == 'G':
        return 6
Titanic['deckR'] = Titanic['deck'].apply(Deck)
Titanic.drop('deck', axis = 1, inplace = True)
Titanic.deckR = Titanic.deckR.astype(int)
```

```python
Titanic.head()
```

```python
Titanic.embarked.value_counts()
```

```python
Titanic.sex.value_counts()
```

```python
def Gender(series):
    if series == 'male':
        return 0
    if series == 'female':
        return 1
Titanic['sexR'] = Titanic['sex'].apply(Gender)
Titanic.drop('sex', axis = 1, inplace = True)
Titanic.sexR = Titanic.sexR.asytpe(int)
```

```python
Titanic.head()
```

```python
y = Titanic.survived
y.head()
```

```python
x = Titanic.drop('survived', axis = 1)
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 76)
```

```python
decisionTree = DecisionTreeClassifier(random_state = 76)
decisionTree.fit (x_train, y_train)
```

```python
treePredictions = decisionTree.predict(x_test)
print(confusion_matrix (y_test, treePredictions))
print (classification_report (y_test, treePredictions))
```

```python
forest = RandomForestClassifier(n_estimators = 500, random_state = 76)
forest.fit (x_train, y_train)
```

```python
forestPredictions = forest.predict(x_test)
print(confusion_matrix(y_test, forestPredictions))
print(classification_report(y_test, forestPredictions))
```

The decision tree model predicts accuracy 76% of the time but the forest model predicts accuracy 80% of the time.  This means that the forest prediction model is the best model to use in this situation.
